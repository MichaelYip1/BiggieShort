```python
# Code to extract quandl api to csv
import quandl
import json
# Initial imports
import numpy as np
import os
import requests
import pandas as pd
from dotenv import load_dotenv
import alpaca_trade_api as tradeapi
from fredapi import Fred
```

```python
# Load .env environment variables
load_dotenv()
```

```python
# Set Fred API Key via 
fred_api = os.getenv('fred_api_key')
fred = Fred(api_key=fred_api)
```

```python
job_market = quandl.get("FRED/NROUST", collapse="quarterly")
job_market_df = pd.DataFrame(job_market,index=job_market.index)
job_market_df.index = pd.to_datetime(job_market_df.index)
#job_market_df.index=job_market_df.index.date

job_market_df.head()
```

```python
gdp = quandl.get("FRED/GDPSOPQ027S", collapse="quarterly")
gdp_df = pd.DataFrame(gdp,index=gdp.index)
gdp_df.index = pd.to_datetime(gdp_df.index)
#gdp_df.index=gdp_df.index.date
gdp_df.tail()
```

```python
inflation = quandl.get("RATEINF/INFLATION_USA", collapse="quarterly")
inflation_df = pd.DataFrame(inflation,index=inflation.index)
inflation_df.index = pd.to_datetime(inflation_df.index)
# inflation_df.index=inflation_df.index.date
#inflation_df = inflation_df.resample('Q').mean()
# inflation_df.tail()
inflation_df
```

```python
irate_data = quandl.get("FED/SVENPY", collapse="quarterly")
irate_data
```

```python
# This to be examined individually
drate_data = {}
drate_data['Delinquency Rates'] = fred.get_series('DRSFRMACBS')
drate_data = pd.DataFrame(drate_data)
drate_data.index = pd.to_datetime(drate_data.index)
drate_data = drate_data.resample('Q').mean()
drate_data = drate_data.dropna()
drate_data
```

```python
price_data = quandl.get("FMAC/HPI_USA", collapse="quarterly")
price_data
```

```python
df = {}
df['Homeownership Rate in the United States'] = fred.get_series('RHORUSQ156N')
# data index not accurate - produces NaN
df['Monthly Supply of Houses in the United States'] = fred.get_series('MSACSR')
df['US Population data'] = fred.get_series('POPTHM')
# data index not accurate - produces NaN 
df['Median Sales Price of Houses Sold'] = fred.get_series('MSPUS')
df = pd.DataFrame(df)
```

```python
df = pd.DataFrame(df,index=df.index)
df.index = pd.to_datetime(df.index)
# .asfreq() instead of .mean()
df = df.resample('Q').mean()
df = df.dropna()
df
```

```python
bot_features = pd.concat([job_market_df,gdp_df,inflation_df,irate_data, price_data,df], axis='columns', join='outer')
bot_features = pd.DataFrame(bot_features,index=bot_features.index)
bot_features.index = pd.to_datetime(bot_features.index)
bot_features = bot_features.resample('Q').mean()
bot_features = bot_features.dropna()
bot_features
```

```python
bot_features.to_csv('bot_features.csv')
```
