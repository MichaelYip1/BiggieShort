### Data clean

1. Job market - https://www.quandl.com/data/FRED/NROUST-Natural-Rate-of-Unemployment-Short-Term
2. GDP - https://www.quandl.com/data/FRED/GDPSOPQ027S-Gross-domestic-product-GDP-sum-of-pieces-Flow
3. Inflation - https://www.quandl.com/data/RATEINF/INFLATION_USA-Inflation-YOY-USA


```python
# Initial imports
import quandl
import pandas as pd
import numpy as np
```

```python
job_market = quandl.get("FRED/NROUST", collapse="quarterly")
job_market_df = pd.DataFrame(job_market,index=job_market.index)
job_market_df.index = pd.to_datetime(job_market_df.index)
#job_market_df.index=job_market_df.index.date

job_market_df.head()
```

```python
gdp = quandl.get("FRED/GDPSOPQ027S", collapse="quarterly")
gdp_df = pd.DataFrame(gdp,index=gdp.index)
gdp_df.index = pd.to_datetime(gdp_df.index)
#gdp_df.index=gdp_df.index.date
gdp_df.tail()
```

```python
inflation = quandl.get("RATEINF/INFLATION_USA", collapse="quarterly")
inflation_df = pd.DataFrame(inflation,index=inflation.index)
inflation_df.index = pd.to_datetime(inflation_df.index)
# inflation_df.index=inflation_df.index.date
#inflation_df = inflation_df.resample('Q').mean()
# inflation_df.tail()
inflation_df
```

```python
df_1 = pd.concat([job_market_df, gdp_df, inflation_df], axis='columns', join='outer')
#df_1=df_1.dropna()
df_1.reset_index().head()
col_names = ["Job Market", "GDP", "Inflation"]
df_1.columns = col_names

df_1.dropna()
```
